#### if we already know that the current changes are the ones that we want to commit;
we can skip the staging step and go directly to the commit though **"git commit -a"**<br>
It seems to be a shortcut, but actually it doesn't commit "untracked files", applicable only to tracked files.<br>
We'll also use the -m flag to add the commit message directly. -> **git commit -a -m "message"**


### Getting more information about changes
We've seen how git log shows us the list of commits made in the current Git repository. <br>
By default, it prints the commit message, the author, and the date of the change. <br>
But if we're combing through a history of changes in a repo to try and find what caused the latest outage,<br>
we'll probably also need to look at the actual lines that changed in each commit.<br> 
To do this with git log, we can use the -p flag. <br>

```python
!git log -p 
```

Using this option, we can quickly see what changes were made to the files in our repository. 

#### If we don't want to scroll down until we find the commit that we're actually interested in; 
another option is to use the **git show** command. This command takes a commit ID as a parameter, and will display the information about the commit and the associated patch. 

```python
!git log
```
